# Bootstrap

```python
from sklearn.utils import resample
import numpy as np
```

```python
#データ
data_x = np.random.normal(0,1,10000)
#bootrap replicationの回数 =>1000~2000
K=2000
#resamplingする後のデータサイズ
s=len(data_x)
```

```python
resampled_data=[]
for i in range(K):
    resampled_data.append(resample(data_x, n_samples = s))
print(resampled_data)
```

```python
resampled_ave=[]
for i in range(K):
    resampled_ave.append(sum(resampled_data[i])/len(resampled_data[i]))
print(resampled_ave)
```

```python
pred_ave=0
for i in range(K):
    pred_ave+=resampled_ave[i]
pred_ave=pred_ave/K
print(pred_ave) #true_aveへの近似
```

```python
true_ave=np.mean(data_x)
true_ave #真の期待値への近似
```

```python
pred_var=0
for i in range(K):
    pred_var += (resampled_ave[i]-pred_ave)**2
pred_var=pred_var/(K-1)
print(pred_var)
pred_sd=pred_var**(1/2)
print(pred_sd)
```

```python
true_var=0
for i in range(len(data_x)):
    true_var+=(data_x[i]-true_ave)**2
true_var=true_var/(len(data_x)-1) # 正規分布の近似
true_sd=true_var**(1/2)
print(true_var)
print(true_sd)
```
